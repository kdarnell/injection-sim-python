```python
import lookup_table as lt
import numpy as np
import dill as pickle
import os
suffix = '_data_dict.pkl'
P_get = 110.0
all_files = os.listdir(os.getcwd())
file_list = [f for f in all_files if suffix in f]
lookup_dict = {}
lt_flash_dict = {}
for filename in file_list:
    P = np.float(filename.split('bar')[0])
    T = np.float(filename.split('_')[1].split('C')[0])
    if P == P_get:
        l_table = lt.quick_lookup(pickle.load(open(filename, 'rb')))
        lookup_dict[T] = {'fn': filename,
                          'P': P,
                          'flash': l_table.flash,
                          'lt': l_table}
```

```python
def gradient_t_flash(z, T_q, lookup_dict):
    lookup_sorted = sorted(lookup_dict.keys())
    ind_q = np.interp(T_q, lookup_sorted, range(len(lookup_dict)))
    ind_set = [int(np.floor(ind_q)), int(np.ceil(ind_q))]
    t_interp_results = [lookup_dict[lookup_sorted[ii]]['flash'](z) for ii in ind_set]
    if (t_interp_results[0][0] == t_interp_results[1][0]) and (ind_set[0] != ind_set[1]):
        comp = t_interp_results[0][1][0] + (t_interp_results[1][1][0] - t_interp_results[0][1][0]) / (ind_set[1] - ind_set[0])*(ind_q - ind_set[0])
        alpha = np.linalg.lstsq(comp, z)[0]
        output = (t_interp_results[0][0], (comp, alpha, None, None), None)
    elif (ind_set[0] == ind_set[1]):
        output = t_interp_results[0]
    else:
        output = lookup_dict[lookup_sorted[int(np.round(ind_q))]]['flash'](z)
    return output
```

```python
gradient_t_flash(np.asarray([0.98, 0.02, 0.0, 0]), 18.55, lookup_dict)
```
