```python
plt.figure()
S = sim_results['results']['S'][:,[0,2,4]]
plt.fill_between(sim_results['xc'], S[:,1] + S[:,2], 1)
plt.fill_between(sim_results['xc'], S[:,1])
plt.fill_between(sim_results['xc'], S[:,1] + S[:,2], S[:,1])
plt.xlabel('Distance, $\\xi$')
plt.ylabel('Saturation, $S$')
plt.legend(['Aqueous phase', 'Vapor phase', 'Hydrate (s2) phase'])
plt.xlim((0,1))
plt.ylim((0,1))
plt.grid('on')
```

```python
plt.figure()
z = sim_results['results']['z']
z_gst = z[:,1:-1] / np.sum(z[:, 1:-1], axis=-1)[:, np.newaxis]
plt.fill_between(sim_results['xc'], z_gst[:,1], 1)
plt.fill_between(sim_results['xc'], z_gst[:,1])
plt.xlabel('Distance, $\\xi$')
plt.ylabel('Mole fraction of guests, $z_i$')
plt.legend(['CH$_4$', 'C$_2$H$_6$'])
plt.xlim((0,1))
plt.ylim((0,1))
plt.grid('on')
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 1)
S = sim_results['results']['S'][:,[0,2,4]]
plt.fill_between(sim_results['xc'], S[:,1] + S[:,2], 1)
plt.fill_between(sim_results['xc'], S[:,1])
plt.fill_between(sim_results['xc'], S[:,1] + S[:,2], S[:,1])
plt.xlabel('Distance, $\\xi$')
plt.ylabel('Saturation, $S$')
plt.legend(['Aqueous phase', 'Vapor phase', 'Hydrate (s2) phase'])
plt.xlim((0,1))
plt.ylim((0,1))
# plt.grid('on')

mpl.rcParams['axes.prop_cycle'] = cycler(color=('#8c564b', '#c5b0d5'))
plt.subplot(2, 2, 2)
V_z = sim_results['results']['x'][:,:-1,2]
V_guest = V_z[:, 1:] / np.sum(V_z[:, 1:], axis=-1)[:, np.newaxis]
plt.plot(sim_results['xc'], V_guest, linewidth=3)
plt.ylabel('Mole fraction in Vapor, $x_V$')
plt.legend(['CH$_4$', 'C$_2$H$_6$'])
plt.xlim((0,1))
plt.ylim((0,1))
plt.grid('on')

plt.subplot(2, 2, 4)
H_z = sim_results['results']['x'][:,:-1,4]
H_guest = H_z[:, 1:] / (np.sum(H_z[:, 1:], axis=-1) + 1e-20)[:, np.newaxis]
H_guest = H_guest * (sim_results['results']['S'][:, 4] >0.01 )[:, np.newaxis]
plt.plot(sim_results['xc'], H_guest, linewidth=3)
plt.ylabel('Mole fraction in Hydrate, $x_H$')
plt.xlabel('Distance, $\\xi$')
plt.legend(['CH$_4$', 'C$_2$H$_6$'])
plt.xlim((0,1))
plt.ylim((0,1))
plt.grid('on')

mpl.rcParams['axes.prop_cycle'] = cycler(color=('#9edae5', '#8c564b', '#c5b0d5'))
plt.subplot(2, 2, 3)
plt.plot(sim_results['xc'], sim_results['results']['z'][:,:-1], linewidth=3)
plt.xlabel('Distance, $\\xi$')
plt.ylabel('Mole fraction, $z$')
plt.legend(['H$_2$O', 'CH$_4$', 'C$_2$H$_6$'])
plt.xlim((0,1))
plt.ylim((0,1))
plt.grid('on')


plt.savefig('test.svg')
plt.show()
```
